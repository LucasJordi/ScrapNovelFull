#Listar todas as obras com base em filtro

```python
filtro="*"
general="https://novelfull.com/search?keyword="+filtro


data=[]
def getData(page):
  r=general+"&page="+str(page)
  wd.get(r)
  content=wd.find_element(By.ID,"list-page")
  novels=content.find_elements(By.XPATH,"//*[@id='list-page']/div[1]/div/div[@class='row']")
  i=1
  print(len(novels))
  
  for var in novels:
    title=var.find_element(By.TAG_NAME,"a")   
    i=i+1
    try:
      cover=var.find_element(By.XPATH,"//*[@id='list-page']/div[1]/div/div["+str(i)+"]/div[1]/div/img").get_attribute("src")
    except:
      print("erro")
    data.append({"id":len(data)+1,"title":title.get_attribute("title"),"link":title.get_attribute("href"),"cover":cover})
    


for i in range(2):
  
  getData(i+1)


print(data)
```
